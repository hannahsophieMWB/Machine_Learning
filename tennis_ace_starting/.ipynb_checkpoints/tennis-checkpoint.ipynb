### Identifying the Data

`Player`: name of the tennis player

`Year`: year data was recorded

**Service Game Columns (Offensive)**

`Aces`: number of serves by the player where the receiver does not touch the ball

`DoubleFaults`: number of times player missed both first and second serve attempts

`FirstServe`: % of first-serve attempts made

`FirstServePointsWon`: % of first-serve attempt points won by the player

`SecondServePointsWon`: % of second-serve attempt points won by the player

`BreakPointsFaced`: number of times where the receiver could have won service game of the player

`BreakPointsSaved`: % of the time the player was able to stop the receiver from winning service game when they had the chance

`ServiceGamesPlayed`: total number of games where the player served

`ServiceGamesWon`: total number of games where the player served and won

`TotalServicePointsWon`: % of points in games where the player served that they won


**Return Game Columns (Defensive)**

`FirstServeReturnPointsWon`: % of opponents first-serve points the player was able to win

`SecondServeReturnPointsWon`: % of opponents second-serve points the player was able to win

`BreakPointsOpportunities`: number of times where the player could have won the service game of the opponent

`BreakPointsConverted`: % of the time the player was able to win their opponent‚Äôs service game when they had the chance

`ReturnGamesPlayed`: total number of games where the player‚Äôs opponent served

`ReturnGamesWon`: total number of games where the player‚Äôs opponent served and the player won

`ReturnPointsWon`: total number of points where the player‚Äôs opponent served and the player won

`TotalPointsWon`: % of points won by the player


**Outcomes**

`Wins`: number of matches won in a year

`Losses`: number of matches lost in a year

`Winnings`: total winnings in USD in a year

`Ranking`: ranking at the end of year

### Correlation Analysis
The correlation coefficient is the specific measure that quantifies the strength of the linear relationship between two variables in a correlation analysis. 

We will use 'pandas.DataFrame.corr' with pearson  standard correlation coefficient. Karl Pearson's coefficient of correlation is defined as a linear correlation coefficient that falls in the value range of -1 to +1. Value of -1 signifies strong negative correlation while +1 indicates strong positive correlation. 

Here we are interested in the correlation coefficient values relative to the columns `Wins`, `Winnings` and `Ranking` to see if strong relationships exists between these outcomes and the features.

We can distguish the correlation table into *Offense* and *Defense*.

```python
# Coefficient of Correlation Matrix DataFrame I
tennis_corr = data.corr() 
#print(tennis_corr)


# Outcome vs features Correlation DataFrame
tennis_corr_outcome = tennis_corr[['Winnings', 'Wins', 'Ranking']].reset_index()
tennis_corr_outcome.rename(columns={'index':'Feature'}, inplace=True)

# Dropping the features that are not important for the analysis
tennis_corr_outcome = tennis_corr_outcome[
                                                       (tennis_corr_outcome.Feature != 'Wins')
                                                      & (tennis_corr_outcome.Feature != 'Losses')
                                                      & (tennis_corr_outcome.Feature != 'Winnings')
                                                      & (tennis_corr_outcome.Feature != 'Ranking')]
# Sort
#tennis_corr_outcome.sort_values(['Winnings'], ascending=False)

# We can break down the list of features into 'Offense' and 'Defense' 
# Note: We take out Year for now and will come back to that later

# Offense Correlation DataFrame
tennis_corr_outcome_off = tennis_corr_outcome.loc[(tennis_corr_outcome.Feature == 'Aces')
                                                      | (tennis_corr_outcome.Feature == 'DoubleFaults')
                                                      | (tennis_corr_outcome.Feature == 'FirstServePointsWon')
                                                      | (tennis_corr_outcome.Feature == 'SecondServePointsWon')
                                                      | (tennis_corr_outcome.Feature == 'BreakPointsFaced')
                                                      | (tennis_corr_outcome.Feature == 'BreakPointsSaved')
                                                      | (tennis_corr_outcome.Feature == 'ServiceGamesWon')
                                                      
                                                      | (tennis_corr_outcome.Feature == 'TotalServicePointsWon')]\
                                                     .sort_values(by='Wins', ascending=False)\
                                                    .reset_index(drop=True)

# Sort
#tennis_corr_outcome_off.sort_values(['Winnings'], ascending=False)

# Defense Correlation DataFrame
tennis_corr_outcome_def = tennis_corr_outcome.loc[(tennis_corr_outcome.Feature == 'FirstServeReturnPointsWon')
                                                      | (tennis_corr_outcome.Feature == 'SecondServeReturnPointsWon')
                                                      | (tennis_corr_outcome.Feature == 'BreakPointsOpportunities')
                                                      | (tennis_corr_outcome.Feature == 'BreakPointsConverted')
                                                      | (tennis_corr_outcome.Feature == 'ReturnGamesWon')
                                                      | (tennis_corr_outcome.Feature == 'ReturnPointsWon')
                                                      | (tennis_corr_outcome.Feature == 'TotalPointsWon')]\
                                                     .sort_values(by='Wins', ascending=False)\
                                                     .reset_index(drop=True)
#tennis_corr_outcome_def.sort_values(['Winnings'], ascending=False)
```

We can see that a mix of *Offense* and *Defense* techniques (`ServiceGamesPlayed`, `ReturnGamesPlayed`, `BreakPointsOpportunities`) etc have a high correlation with Winnings and Wins.

## Break Points

During the exploratory analysis, we can see a strong correlation between `BreakPointsFaced`,  `BreakPointsOpportunities`  and winnings. This is exciting so let us dive a little bit deeper in the relationship between the offensive and defensive tactics here. 

So far we only know that: 

- More Break Points Opportunities (defensive) and Break Points Faced (offensive) meant more winnings.

A break point opportunity occurs when two conditions are met: 

- 1) one player is one point away from a win, and 
- 2) their opponent is serving. 

Essentially, in order to win the game ‚Äî and inch closer to winning the set and match ‚Äî the player must "break" their opponent's serve. Which essentially means, he must convert the  `Break Points Opportunities`. 

We are interested in the rate that the receiving player  would need to  convert `Break Points Opportunities`  to win the game.


Let‚Äôs move to the server‚Äôs side of the net and take a look at `BreakPointsFaced`. This is when a server is one point away from losing a game, but are able to make a comeback to prevent the receiver from securing a break point:

We will plot Break Points  as a scatterplot against `Winnings` in order to find out what conversation rate a player should aim for.


```python
break_points_opportunities = np.array(data.BreakPointsOpportunities).reshape(-1,1)
break_points_faced = np.array(data.BreakPointsFaced).reshape(-1,1)
winnings = np.array(data.Winnings).reshape(-1,1)
bp_predict = [[i] for i in range(0,600)]

bpo_lr = LinearRegression()
bpo_lr.fit(break_points_opportunities, winnings)
bpo_predict = bpo_lr.predict(bp_predict)

bpf_lr = LinearRegression()
bpf_lr.fit(break_points_faced, winnings)
bpf_predict = bpf_lr.predict(bp_predict)
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Break Points vs. Winnings')

plt.subplot(221)
plt.scatter(data.BreakPointsConverted, data.Winnings, alpha=0.4)
plt.xlabel('Break Points Converted')
plt.ylabel('Winnings')

plt.subplot(222)
plt.scatter(data.BreakPointsSaved, data.Winnings, alpha=0.4)
plt.xlabel('Break Points Saved')
plt.ylabel('Winnings')

plt.subplot(223)
plt.scatter(data.BreakPointsOpportunities, data.Winnings, alpha=0.4)
plt.plot(bp_predict, bpo_predict, color='orange')
plt.xlabel('Break Points Opportunities')
plt.ylabel('Winnings')
plt.ylim(0,1200000)
plt.xlim(0,600)

plt.subplot(224)
plt.scatter(data.BreakPointsFaced, data.Winnings, alpha=0.4)
plt.plot(bp_predict, bpf_predict, color='orange')
plt.xlabel('Break Points Faced')
plt.ylabel('Winnings')
plt.ylim(0,1200000)
plt.xlim(0,600)

plt.subplots_adjust(wspace=0.4)

plt.show()
```

### Do Break Points Really Matter?
It‚Äôs easy to see why break points could simply be viewed as ‚Äòjust another point.‚Äô However, the fact of the matter is that the result of break points can significantly alter the course of a match.

We knew that winnings were higher the more `Break Points Opportunities` a player converted and `Break Points Faced` a player saved.

The 2 top graphs display that very few tennis players converted or saved 100% of their break points. 

We can see that a player was more likely to make a big buck  if she or he converted around 40% and saved around 60% of break points.

Becasue the conversion/saving rate (40%/60%) showing a higher correlation with winnings, a player who would play longer games with more break points would bring home more money. 

As a result, we could go to a tennis coach and advice him to train his player to go for the following:

- When playing defence: Convert about 40% of Break Points to achieve  higher earnings at this frequency.
- When playin offense: Save about 60% of Break Points to achieve  higher earnings at this frequency.




# ‚ú®  Regression models

We use a set of regression models to predict the outcomes.

We will show how the best top four correlation features (`BreakPointsOpportunities`, `BreakPointsFaced`, `DoubleFaults` and `Aces`)  predict `Winnings`. 

Before training the model, it is important to split our data into training and test datasets so that we can evaluate our model on the test set. Let's split the data set into training and evaluating sets using train_test_split() method. 80% of the data will serve as a training set, and 20% of the data will be testing set.



### Feature selection for regression model 

We select the features with the highest correlation with our target variable. 

```python
# Best features 

features_best = tennis_corr_outcome.loc[(tennis_corr_outcome.Feature == 'Aces')
                                                      | (tennis_corr_outcome.Feature == 'BreakPointsFaced')
                                                      | (tennis_corr_outcome.Feature == 'BreakPointsOpportunities')
                                                      | (tennis_corr_outcome.Feature == 'DoubleFaults')] \
                                                        .sort_values(by='Wins', ascending=False)\
                                                     .reset_index(drop=True)

features_best
                                        

```

```python
# Target outcome

y = data[['Winnings']]

# Features name list
features = ['BreakPointsOpportunities', 'Aces', 'BreakPointsFaced', 'DoubleFaults' ]

x = data[features]
```

```python
# Split the data

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)

# Regressor
lm = LinearRegression()

# Fit the model
model = lm.fit(x_train, y_train)

y_predict= lm.predict(x_test)


print("Train score:")
print(lm.score(x_train, y_train))

print("Test score:")
print(lm.score(x_test, y_test))

#Visualise the results
plt.scatter(y_test, y_predict)
plt.plot(range(1100000), range(1100000), color='orange')
plt.title('Actual Winnings vs Predicted Winnings')
plt.ylim(0,1100000)
plt.xlim(0,1100000)

plt.show()
```

```python
# Evaluate the model
X2 = sm.add_constant(x)
est = sm.OLS(y, X2)
reg = est.fit()

pred = reg.predict(X2)

print(reg.summary())

predictions_tennis = pred
```

From the regression model results, we can interpret a couple of the indicators:
- **p-value**
The p-value for each independent variable tests the null hypothesis that the feature has no correlation with the dependent variable (`Winnings`). If there is no correlation, there is no association between the changes in the independent variable and the shifts in the dependent variable. In other words, there is insufficient evidence to conclude that there is an effect at the population level. For the following features, we infer that we do have a significant correlation with `Winnings`:

    - `BreakPointsOpportunities` 
    - `Aces` 
    - `DoubleFaults`

Only `BreakPointsFaced` seems to be non-significant which makes sense since it is a difficult situation in which the defence player is facing a break-point opportunity and he or she would not want to have too many of these situations.

- **coefficients**
The regression coefficient tells us whether we have a positive or negative correlation between each independent feature and the target  variable  (`Winnings`). A positive coefficient indicates that as the value of the independent variable increases, the mean of the dependent variable also tends to increase. A negative coefficient suggests that as the independent variable increases, the dependent variable tends to decrease.

The coefficient value signifies how much the mean of the dependent variable changes given a one-unit shift in the independent variable while holding other variables in the model constant. We can infer that each additional  `BreakPointsOpportunity` would lead to a increase in Winnings by USD  1427 on average. Pretty hot stuff. 

- **R-squared**
This is our goodness-of-fit measure for linear regression models. This statistic indicates the percentage of the variance in the dependent variable that the independent variables explain collectively. R-squared measures the strength of the relationship between your model and the dependent variable on a convenient 0 ‚Äì 100% scale. R-squared is the percentage of the dependent variable variation that a linear model explains.
After fitting our regression model, we can determine that our model fits the data pretty well (83%).
